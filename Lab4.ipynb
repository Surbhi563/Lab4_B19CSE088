# Foodgrains Stock Visualisation

```python
food=pd.read_csv('/content/Production_of_Agriculture_Stock-Ministry_of_Finance.csv')
food.drop(['Frequency','Unit'],axis=1,inplace=True)
food= food.T
food1=food.reset_index(inplace=True)
food.columns=food.iloc[0]
food=food[1:]
food
```

```python
import pandas as pd
import matplotlib.pyplot as plt

l=food.columns
# l.remove('Particulars')
plt.figure(figsize=(15, 10))
for i in l:
  if i!='Particulars':
    plt.plot(food['Particulars'],food[i],label=i)
plt.xlabel('Months')
plt.ylabel('Quantity (in million tonnes)')
plt.legend()
plt.show()
```

# Colleges visualisation

```python
colleges=pd.read_csv('/content/MergerdAyush_Colleges_in_India_01-04-2021_0.csv')
colleges.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# create a pandas dataframe from the table
df =colleges

# filter out the columns for Ayurveda category and state-wise college count
df_ayurveda = df[df["Type of System"] == "Ayurveda"][["State or Union Territory", "No. of Colleges - Govt", "No. of Colleges - Non-Govt"]]

# set the state name as the index of the dataframe
df_ayurveda.set_index("State or Union Territory", inplace=True)

# plot a bar chart of the state-wise college count for Ayurveda category
ax = df_ayurveda.plot(kind="bar", figsize=(16, 8), width=0.8, color=["orange", "blue"])

# set the title and axis labels of the plot
ax.set_title("State-wise Colleges of Ayurveda Category", fontsize=20)
ax.set_xlabel("State or Union Territory", fontsize=16)
ax.set_ylabel("Number of Colleges", fontsize=16)

# show the plot
plt.show()
```

# Road transport visualisation

```python
trans=pd.read_csv('/content/Road_Transport_BRS_2017-18_Annexure7_9c.csv')
trans.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# read the dataset into a pandas dataframe
df =trans

# create a stacked bar chart to show the distribution of different types of roads for each state
fig, ax = plt.subplots(figsize=(12,6))
df.plot(kind='bar', x="Name of the States", y=['National Highways', 'State Highways', 'District Roads'], stacked=True, ax=ax)

# set the chart title and axis labels
ax.set_title('Distribution of National, State, and District Roads by State')
ax.set_xlabel('State')
ax.set_ylabel('Road Length')

# display the chart
plt.show()
```

# Screen Display

```python
import pandas as pd
import matplotlib.pyplot as plt
df2=colleges
# Create a figure with 3 subplots
fig, axs = plt.subplots(1, 3, figsize=(6.4,4.8),dpi=80)

# Subplot 1: Monthly changes in different food grain stocks
for i in l:
  if i!='Particulars':
    axs[0].plot(food['Particulars'],food[i],label=i)
axs[0].set_xlabel('Month')
axs[0].set_ylabel('Stock (in 000 MT)')
axs[0].set_title('Monthly changes in food grain stocks')
axs[0].legend()

# Subplot 2: State-wise colleges of different Ayus categories
df2_grouped = df2.groupby('State or Union Territory')['Type of System'].count()
df2_grouped.plot(kind='bar', stacked=True, ax=axs[1])
axs[1].set_xlabel('State')
axs[1].set_ylabel('Number of colleges')
axs[1].set_title('State-wise colleges of different Ayus categories')

# Subplot 3: Distribution of National, State, and District roads
df=trans
# calculate the road length per 1000 population for each category
df['NH_1000_pop'] = df['National Highways'] / (df['Total Population'] / 1000)
df['SH_1000_pop'] = df['State Highways'] / (df['Total Population'] / 1000)
df['DR_1000_pop'] = df['District Roads'] / (df['Total Population'] / 1000)
df['RR_1000_pop'] = df['Rural Roads'] / (df['Total Population'] / 1000)
df['UR_1000_pop'] = df['Urban Roads'] / (df['Total Population'] / 1000)
df['PR_1000_pop'] = df['Project Roads'] / (df['Total Population'] / 1000)

# plot the data
axs[2].plot(df['Name of the States'], df['NH_1000_pop'], label='National Highways')
axs[2].plot(df['Name of the States'], df['SH_1000_pop'], label='State Highways')
axs[2].plot(df['Name of the States'], df['DR_1000_pop'], label='District Roads')
axs[2].plot(df['Name of the States'], df['RR_1000_pop'], label='Rural Roads')
axs[2].plot(df['Name of the States'], df['UR_1000_pop'], label='Urban Roads')
axs[2].plot(df['Name of the States'], df['PR_1000_pop'], label='Project Roads')

axs[2].set_ylabel('Road length per 1000 population')
plt.legend()
plt.show()


```
